# The Data:
We decided to collect three seperate data tables from the "Personal Genome Project" site (located at https://my.pgp-hms.org/google_surveys). The participant IDs are consistent throughout the varying files, allowing comorbidities in each subset to be easily compared. We chose to use the 2012 Nervous System survey (with IDs and conditions), the 2012 Circulatory System survey (with IDs and conditions), and the 2012 Endocrine, Metabolic, Nutritional, and Immunity survey (with IDs and conditions).

Although there are a number of cool datasets on this site, we chose these three because they incorperated or directly impacted the progression of migraines [migraines are a nervous system disorder, migraines lead to changes in blood circulation, and migraines are drastically impacted by fluctuating hormones]. Later on, we hope to expand to other datasets that may be relevant for wider crossreferencing.

```python
Nerv_data = pd.read_csv("/content/drive/MyDrive/Colab_Notebooks/PGPTrait&DiseaseSurvey2012_NervousSystem-20181010220056 (1).csv") # Participant IDs and Nervous system conditions
Circ_data = pd.read_csv("/content/drive/MyDrive/Colab_Notebooks/PGPTrait&DiseaseSurvey2012_CirculatorySystem-20181010220109.csv") # Participant IDs and Circulatory system conditions
Endo_data = pd.read_csv("/content/drive/MyDrive/Colab_Notebooks/PGPTrait&DiseaseSurvey2012_Endocrine,Metabolic,Nutritional,AndImmunity-20181010220044.csv") # Participant IDs and Endocrine system Conditions
Nerv_data.head()
```

```python
Nerv_data = Nerv_data.drop("Do not touch!", axis=1) #dropping unncessary columns
Nerv_data = Nerv_data.drop("Timestamp", axis=1)
Nerv_data = Nerv_data.drop("Other condition not listed here?", axis=1)
Circ_data = Circ_data.drop("Do not touch!", axis=1)
Circ_data = Circ_data.drop("Timestamp", axis=1)
Circ_data = Circ_data.drop("Other condition not listed here?", axis=1)
Endo_data = Endo_data.drop("Do not touch!", axis=1)
Endo_data = Endo_data.drop("Timestamp", axis=1)
Endo_data = Endo_data.drop("Other condition not listed here?", axis=1)
```

```python
Nerv_data.columns=["Participant ID", "Nervous System Conditions"] # distinguishing column names
Circ_data.columns=["Participant ID", "Circulatory System Conditions"]
Endo_data.columns=["Participant ID", "Endocrine System Conditions"]
```

```python
Nerv_Circ_data = Nerv_data.merge(Circ_data, on="Participant ID", how="inner")
full_data = Nerv_Circ_data.merge(Endo_data, on="Participant ID", how="inner")
full_data
```

```python
full_data.info()
```

```python
mig_w_aura = full_data['Nervous System Conditions'].str.contains('Migraine with aura', case=False, na=False) # Any participants with "Migraines with aura" results as True
w_sum = mig_w_aura.sum() # Sum of True values for "Migraines with aura"

mig_no_aura = full_data['Nervous System Conditions'].str.contains('Migraine without aura', case=False, na=False) # Any participants with "Migraines without aura" results as True
no_sum = mig_no_aura.sum() # Sum of True values for "Migraines without aura"

rays = full_data['Circulatory System Conditions'].str.contains("Raynaud's phenomenon", case=False, na=False) # Any participants with "Raynaud's Phenomenon" results as True
rays_sum = rays.sum() # Sum of True values for "Raynaud's Phenomenon"

hypo = full_data['Endocrine System Conditions'].str.contains("Hypothyroidism", case=False, na=False) # Any participants with "Hypothyroidism" results as True
hypo_sum = hypo.sum() # Sum of True values for "Hypothyroidism"

ray_mig_no = ((rays) & (mig_no_aura)) # Filters for participants with Raynaud's and Migraines without aura
ray_mig_no_sum = ray_mig_no.sum() # sum of participants with Raynaud's and Migraines without aura

ray_mig_w = ((rays) & (mig_w_aura)) # Filters for participants with Raynaud's and Migraines with aura
ray_mig_w_sum = ray_mig_w.sum() # sum of participants with Raynaud's and Migraines with aura

hypo_mig_w = ((hypo) & (mig_w_aura)) # Filters for participants with Hypothyroidism and Migraines with aura
hypo_mig_w_sum = hypo_mig_w.sum() # sum of participants with Hypothyroidism and Migraines with aura

hypo_mig_no = ((hypo) & (mig_no_aura)) # Filters for participants with Hypothyroidismf and Migraines without aura
hypo_mig_no_sum = hypo_mig_no.sum() # sum of participants with Hypothyroidism and Migraines without aura

hypo_mig_w_ray = ((hypo_mig_w) & (rays)) # Filters for participants with Raynaud's and Migraines with aura and Hypothyroidism
hypo_mig_w_ray_sum = hypo_mig_w_ray.sum() # sum of participants with Raynaud's and Migraines with aura and Hypothyroidism

hypo_mig_no_ray = ((hypo_mig_no) & (rays)) # Filters for participants with Raynaud's and Migraines without aura and Hypothyroidism
hypo_mig_no_ray_sum = hypo_mig_no_ray.sum() # sum of participants with Raynaud's and Migraines without aura and Hypothyroidism
```

```python
plot_list = [w_sum,no_sum,rays_sum,hypo_sum,hypo_mig_w_sum,hypo_mig_no_sum,hypo_mig_w_ray_sum,hypo_mig_no_ray_sum] 
labels = ["w_sum","no_sum","rays_sum","hypo_sum","hypo_mig_w_sum","hypo_mig_no_sum","hypo_mig_w_ray_sum","hypo_mig_no_ray_sum"]
plt.figure(figsize=(18,7))
plt.bar(labels,plot_list)
```
